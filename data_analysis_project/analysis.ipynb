```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
TwitchDataSet = pd.read_csv("../DATA/TwitchDataSet.csv")
```

```python
TwitchDataSet.info()
```

```python
TwitchDataSet = TwitchDataSet.dropna()
plt.subplots(figsize=(28,10))
sns.countplot(data=TwitchDataSet, x="language", order = TwitchDataSet['language'].value_counts().index)
```

```python
result = TwitchDataSet.groupby("first category").sum().sort_values("watch time", ascending=False).head(10)
result.reset_index(inplace=True)
sns.catplot(data=result, y="first category", x="watch time", kind="bar", height=20)
```
